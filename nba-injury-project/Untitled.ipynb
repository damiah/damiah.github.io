```python
from nba_api.stats.endpoints import playercareerstats, commonplayerinfo
from nba_api.stats.static import players, teams
import pandas as pd
import time
```

```python
#We need a separate call for playoffs and regular season
hist_player_log = playercareerstats.PlayerCareerStats(player_id='202695',
                                                  per_mode36='PerGame').get_data_frames()[0]
hist_player_log.head()
```

```python
nba_players = players.get_players()
print('Number of players fetched: {}'.format(len(nba_players)))
nba_teams = teams.get_teams()
print('Number of teams fetched: {}'.format(len(nba_teams)))
```

```python
#Grab all player-season combos
hist_player_full = pd.DataFrame()
nba_players[0]['full_name']
for i in range(0, len(nba_players)):
    player_id = nba_players[i]['id']
    hist_player = playercareerstats.PlayerCareerStats(player_id=player_id, 
                                                  per_mode36='PerGame'
                                                  ).get_data_frames()[0]
    hist_player['full_name'] = nba_players[i]['full_name']
    hist_player_full = hist_player_full.append(hist_player, ignore_index=True)
    if (i % 100 == 0): 
        time.sleep(0.1)
        print(str(i) + "/" + str(len(nba_players)))
   
```

```python
#Grab player info (date of birth, draft year, etc.)
hist_player_info_full = pd.DataFrame()
for i in hist_player_log_full['Player_ID'].unique():
    player_id = i
    player_info = commonplayerinfo.CommonPlayerInfo(player_id=player_id).get_data_frames()[0][
        ['PERSON_ID', 'BIRTHDATE', 'FROM_YEAR', 'HEIGHT', 'WEIGHT', 'POSITION']]
    player_info['full_name'] = hist_player_log_full['full_name'][hist_player_log_full['Player_ID']==i].unique()[0]
    hist_player_info_full = hist_player_info_full.append(player_info, ignore_index=True)
    time.sleep(0.25)
```
